```python
#data analysis on product inventory using pandas
import pandas as pd

data = {
'Product':['Laptop','Smartphone','Tablet','Monitor','Keyboard'],
'Price':[1200,800,400,300,50],
'Quantity':[10,25,15,8,40]}
df = pd.DataFrame(data)
print("Inventory Data:\n",df)
df["Total_Value"] = df["Price"] * df["Quantity"]
print("\nInventory with Total values")
print(df)
print("\nAverage Price:",df['Price'].mean())
print("\nMaximum Prices:",df['Price'].max())
print("\nMinimum Prices:",df['Price'].min())
```
