```python
#matrix manipulation on stock prize data
import numpy as np
stock1 = np.array([
    [100,110,95],
    [102,112,97],
    [105,115,100],
    [108,118,102],
    [110,120,105]])
stock2 = np.array([
    [98,108,94],
    [101,111,96],
    [103,113,98],
    [107,117,101],
    [109,119,104]])
sum_matrix=np.add(stock1,stock2)
diff_matrix=np.subtract(stock1,stock2)
print("Stock 1 Data:\n",stock1)
print("Stock 2 Data:\n",stock2)
print("Element-wise Sum")
print(sum_matrix)
print("Element-wise Difference")
print(diff_matrix)


```
