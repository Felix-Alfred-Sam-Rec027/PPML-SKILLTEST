```python
# Analyze EEG Data using Pandas

import pandas as pd
import matplotlib.pyplot as plt

# Create EEG dataset
data = {
    'Participant':[1,1,1,2,2,2],
    'Electrode':['E1','E2','E3','E1','E2','E3'],
    'Time':[1,2,3,1,2,3],
    'Value':[10,12,11,13,9,15]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display dataset
print("EEG Data:\n", df)

# Mean values
print("\nMean Values:")
print(df.groupby(['Participant','Electrode'])['Value'].mean())

# Variance
print("\nVariance:")
print(df.groupby(['Participant','Electrode'])['Value'].var())

# Peak values
print("\nPeak Values:")
print(df.groupby(['Participant','Electrode'])['Value'].max())

# Trend analysis
df.groupby('Time')['Value'].mean().plot()

plt.title("Brainwave Activity Over Time")
plt.xlabel("Time")
plt.ylabel("Signal Value")
plt.show()
```
